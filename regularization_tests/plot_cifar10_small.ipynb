```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# Mixer

```python
# mixer_smallset_vanilla
with open('./data_cifar10/mixer_smallset_vanilla.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# mixer_smallset_lc2_frac8
with open('./data_cifar10/mixer_smallset_lc2_frac8.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

# mixer_smallset_lc1_frac5
with open('./data_cifar10/mixer_smallset_lc2_frac5.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc3 = np.array(data['train_accs']) * 100
val_acc3 = np.array(data['val_accs'])

plt.plot(train_acc1, label='vanilla train')
plt.plot(val_acc1, label='vanilla val')

plt.plot(train_acc2, label='lc2_frac8 train')
plt.plot(val_acc2, label='lc2_frac8 val')

plt.plot(train_acc3, label='lc1_frac5 train')
plt.plot(val_acc3, label='lc1_frac5 val')

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()

print('vanilla val acc: ', np.max(val_acc1))
print('lc2_frac8 val acc: ', np.max(val_acc2))
print('lc1_frac5 val acc: ', np.max(val_acc3))
```
