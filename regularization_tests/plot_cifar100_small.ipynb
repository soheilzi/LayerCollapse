```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# VGG16

```python
# vgg_smallset_vanilla
with open('./data_cifar100/vgg_smallset_vanilla.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# vgg_smallset_lc
with open('./data_cifar100/vgg_smallset_lc.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

# vgg_smallset_wd
with open('./data_cifar100/vgg_smallset_wd.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc3 = np.array(data['train_accs']) * 100
val_acc3 = np.array(data['val_accs'])

plt.plot(train_acc1, label='train_acc_vanilla')
plt.plot(val_acc1, label='val_acc_vanilla')

plt.plot(train_acc2, label='train_acc_lc')
plt.plot(val_acc2, label='val_acc_lc')

plt.plot(train_acc3, label='train_acc_wd')
plt.plot(val_acc3, label='val_acc_wd')

plt.legend()
plt.show()

print('val_acc_vanilla: ', np.max(val_acc1))
print('val_acc_lc: ', np.max(val_acc2))
print('val_acc_wd: ', np.max(val_acc3))
```
