```python
# LC1_full_dataset
with open('./data/LC1_full_dataset.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])
```
