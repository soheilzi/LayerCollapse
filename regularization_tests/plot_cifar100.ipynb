```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# Full dataset VGG16

## correct scale

```python
# LC_full_dataset_imagesize32
with open('./data_cifar100/LC_full_dataset_imagesize32.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# noreg_full_dataset_correct_size
with open('./data_cifar100/noreg_full_dataset_correct_size.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

# small_net_full_dataset_imagesize32
with open('./data_cifar100/small_net_full_dataset_imagesize32.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc3 = np.array(data['train_accs']) * 100
val_acc3 = np.array(data['val_accs'])
```

```python
plt.plot(train_acc1, label='train LC')
plt.plot(val_acc1, label='val LC')

plt.plot(train_acc2, label='train noreg')
plt.plot(val_acc2, label='val noreg')

plt.plot(train_acc3, label='train small')
plt.plot(val_acc3, label='val small')

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()
```

## scaled to 224

```python

# small_net_full_dataset
with open('./data_cifar100/small_net_full_dataset.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# noreg_full_dataset
with open('./data_cifar100/noreg_full_dataset.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

# LC_full_dataset
with open('./data_cifar100/LC_full_dataset.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc3 = np.array(data['train_accs']) * 100
val_acc3 = np.array(data['val_accs'])
```

```python
plt.plot(train_acc1, label='train small_net')
plt.plot(val_acc1, label='val small_net')

plt.plot(train_acc2, label='train noreg')
plt.plot(val_acc2, label='val noreg')

plt.plot(train_acc3, label='train LC')
plt.plot(val_acc3, label='val LC')

plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.legend()
plt.show()
```
