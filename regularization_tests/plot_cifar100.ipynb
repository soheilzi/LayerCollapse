```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# Full dataset VGG16

## Mixer

```python
# mixer_vanilla
with open('./data_cifar100/mixer_vanilla.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# dropout_mixer
with open('./data_cifar100/dropout_mixer.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

# dropout_mixer_lc2
with open('./data_cifar100/dropout_mixer_lc2.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc3 = np.array(data['train_accs']) * 100
val_acc3 = np.array(data['val_accs'])

# dropout_mixer_lc3_fraction8
with open('./data_cifar100/dropout_mixer_lc3_fraction8.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc4 = np.array(data['train_accs']) * 100
val_acc4 = np.array(data['val_accs'])

# dropout_mixer_lc2_fraction10
with open('./data_cifar100/dropout_mixer_lc2_fraction10.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc5 = np.array(data['train_accs']) * 100
val_acc5 = np.array(data['val_accs'])


plt.plot(train_acc1, label='mixer_vanilla')
plt.plot(val_acc1, label='mixer_vanilla_val')
plt.plot(train_acc2, label='dropout_mixer')
plt.plot(val_acc2, label='dropout_mixer_val')
plt.plot(train_acc3, label='dropout_mixer_lc2')
plt.plot(val_acc3, label='dropout_mixer_lc2_val')
plt.plot(train_acc4, label='dropout_mixer_lc3_fraction8')
plt.plot(val_acc4, label='dropout_mixer_lc3_fraction8_val')
plt.plot(train_acc5, label='dropout_mixer_lc2_fraction10')
plt.plot(val_acc5, label='dropout_mixer_lc2_fraction10_val')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

print('mixer_vanilla: ', np.max(val_acc1))
print('dropout_mixer: ', np.max(val_acc2))
print('dropout_mixer_lc2: ', np.max(val_acc3))
print('dropout_mixer_lc3_fraction8: ', np.max(val_acc4))
print('dropout_mixer_lc2_fraction10: ', np.max(val_acc5))
```

```python
# timm_vit_lc2_frac8
with open('./data_cifar100/timm_vit_lc2_frac8.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# timm_vit_lc2_frac8_corrected
with open('./data_cifar100/timm_vit_lc2_frac8_corrected.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

plt.plot(train_acc1, label='timm_vit_base')
plt.plot(val_acc1, label='timm_vit_base_val')
plt.plot(train_acc2, label='timm_vit_lc2_frac8_corrected')
plt.plot(val_acc2, label='timm_vit_lc2_frac8_corrected_val')

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

print('timm_vit_base: ', np.max(val_acc1))
print('timm_vit_base: ', np.max(val_acc2))
```
