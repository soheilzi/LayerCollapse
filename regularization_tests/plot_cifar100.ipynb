```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data_cifar100cifar100_vanilla
with open('data_cifar100cifar100_vanilla.pkl', 'rb') as f:
    data = pickle.load(f)
train_accs1 = np.array(data['train_accs']) * 100
val_accs1 = np.array(data['val_accs'])

# data_cifar100cifar100_LC
with open('data_cifar100cifar100_LC.pkl', 'rb') as f:
    data = pickle.load(f)
train_accs2 = np.array(data['train_accs']) * 100
val_accs2 = np.array(data['val_accs'])

```

```python
plt.plot(train_accs1, label='train_accs vanilla')
plt.plot(val_accs1, label='val_accs vanilla')

plt.plot(train_accs2, label='train_accs LC')
plt.plot(val_accs2, label='val_accs LC')

plt.legend()
plt.show()
```
