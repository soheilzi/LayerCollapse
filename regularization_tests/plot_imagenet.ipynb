```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

# Small Set
## Mixer

```python
# mixer_smallset_vanilla
with open('./data_imagenet/mixer_smallset_vanilla.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc1 = np.array(data['train_accs']) * 100
val_acc1 = np.array(data['val_accs'])

# mixer_smallset_lc05_frac8
with open('./data_imagenet/mixer_smallset_lc05_frac8.pkl', 'rb') as f:
    data = pickle.load(f)
train_acc2 = np.array(data['train_accs']) * 100
val_acc2 = np.array(data['val_accs'])

plt.plot(train_acc1, label='train vanilla')
plt.plot(val_acc1, label='val vanilla')

plt.plot(train_acc2, label='train lc05_frac8')
plt.plot(val_acc2, label='val lc05_frac8')

plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Imagenet')
plt.show()
```
